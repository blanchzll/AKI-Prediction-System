```python
# -*- coding: utf-8 -*-
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from scipy.stats import pearsonr, spearmanr
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
import warnings
import os

# 设置绘图风格和字体
plt.rcParams['font.sans-serif'] = ['SimHei', 'Arial']  # 适配中文
plt.rcParams['axes.unicode_minus'] = False
sns.set(style="whitegrid", font='SimHei')
warnings.filterwarnings('ignore')

# 创建输出目录
if not os.path.exists('eda_results'):
    os.makedirs('eda_results')

# =============================================================================
# 第一部分：增强的探索性数据分析 (EDA)
# =============================================================================
def enhanced_eda(data, target_col='Tm of MD (K)'):
    """
    执行全面的探索性数据分析并保存图表
    :param data: 包含特征和目标变量的DataFrame
    :param target_col: 目标变量的列名
    """
    print(f"\n=== 开始增强EDA分析 (目标变量: {target_col}) ===")
    
    # 假设前两列是描述性信息(如ID, Formula)，从第3列开始是特征
    # 如果您的数据全是特征和目标，请调整这里的切片
    feature_data = data.drop(columns=[target_col]).iloc[:, 2:] 
    target_data = data[target_col]
    feature_names = feature_data.columns.tolist()

    # 1. 目标变量分布图
    plt.figure(figsize=(10, 6))
    sns.histplot(target_data, kde=True, bins=20, color='steelblue', edgecolor='black')
    plt.axvline(target_data.mean(), color='red', linestyle='--', label=f'均值: {target_data.mean():.2f}')
    plt.axvline(target_data.median(), color='green', linestyle='--', label=f'中位数: {target_data.median():.2f}')
    plt.title(f'{target_col} 分布直方图', fontsize=14)
    plt.xlabel(target_col)
    plt.ylabel('频数')
    plt.legend()
    plt.tight_layout()
    plt.savefig('eda_results/1_target_distribution.png', dpi=300)
    plt.show()

    # 2. 正态性检验 (Q-Q图)
    plt.figure(figsize=(10, 6))
    stats.probplot(target_data, dist="norm", plot=plt)
    plt.title(f'{target_col} Q-Q图 (正态性检验)', fontsize=14)
    plt.tight_layout()
    plt.savefig('eda_results/2_qq_plot.png', dpi=300)
    plt.show()

    # 3. 特征统计概览热图
    # 计算每个特征的统计量
    feat_stats = feature_data.describe().T[['mean', 'std', 'min', 'max']]
    # 为了可视化效果，进行标准化显示
    feat_stats_norm = (feat_stats - feat_stats.mean()) / feat_stats.std()
    
    plt.figure(figsize=(12, max(6, len(feature_names)*0.3))) # 动态调整高度
    sns.heatmap(feat_stats_norm, annot=True, fmt='.2f', cmap='RdBu_r', center=0)
    plt.title('特征统计分布热图 (标准化后)', fontsize=14)
    plt.tight_layout()
    plt.savefig('eda_results/3_feature_stats_heatmap.png', dpi=300)
    plt.show()

    # 4. 相关性分析 (Pearson & Spearman)
    pearson_list = []
    spearman_list = []
    
    for col in feature_names:
        # 处理可能的NaN
        valid_idx = feature_data[col].notna() & target_data.notna()
        if valid_idx.sum() > 1:
            p_corr, _ = pearsonr(feature_data.loc[valid_idx, col], target_data[valid_idx])
            s_corr, _ = spearmanr(feature_data.loc[valid_idx, col], target_data[valid_idx])
            pearson_list.append(abs(p_corr)) # 取绝对值看相关强度
            spearman_list.append(abs(s_corr))
        else:
            pearson_list.append(0)
            spearman_list.append(0)

    # 绘制Top 10相关性特征
    top_n = min(10, len(feature_names))
    top_idx = np.argsort(pearson_list)[-top_n:]
    
    plt.figure(figsize=(12, 6))
    plt.barh(range(top_n), [pearson_list[i] for i in top_idx], color='#6a0dad', alpha=0.7, label='Pearson (线性)')
    plt.barh(range(top_n), [spearman_list[i] for i in top_idx], color='#ff8c00', alpha=0.5, label='Spearman (非线性)')
    plt.yticks(range(top_n), [feature_names[i] for i in top_idx])
    plt.xlabel('相关系数绝对值')
    plt.title(f'与目标变量相关性最高的 {top_n} 个特征', fontsize=14)
    plt.legend()
    plt.tight_layout()
    plt.savefig('eda_results/4_top_correlations.png', dpi=300)
    plt.show()

    # 5. 特征互相关矩阵 (检查共线性)
    # 只取Top 15特征以避免图表过大
    top_15_idx = np.argsort(pearson_list)[-15:]
    top_15_names = [feature_names[i] for i in top_15_idx]
    
    corr_mat = feature_data[top_15_names].corr()
    mask = np.triu(np.ones_like(corr_mat, dtype=bool))
    
    plt.figure(figsize=(12, 10))
    sns.heatmap(corr_mat, mask=mask, annot=True, fmt='.2f', cmap='coolwarm', center=0, square=True, cbar_kws={"shrink": .5})
    plt.title('Top特征互相关矩阵 (检查多重共线性)', fontsize=14)
    plt.tight_layout()
    plt.savefig('eda_results/5_correlation_matrix.png', dpi=300)
    plt.show()

    # 6. 关键特征散点图 (Top 3)
    top_3_idx = np.argsort(pearson_list)[-3:][::-1] # 降序
    
    plt.figure(figsize=(18, 5))
    for i, idx in enumerate(top_3_idx):
        col = feature_names[idx]
        plt.subplot(1, 3, i+1)
        plt.scatter(feature_data[col], target_data, alpha=0.6, c='teal', edgecolors='w')
        
        # 添加趋势线
        z = np.polyfit(feature_data[col], target_data, 1)
        p = np.poly1d(z)
        plt.plot(feature_data[col], p(feature_data[col]), "r--", lw=2)
        
        plt.xlabel(col)
        plt.ylabel(target_col)
        plt.title(f'{col} vs Target (Corr: {pearson_list[idx]:.2f})')
        plt.grid(True, alpha=0.3)
        
    plt.tight_layout()
    plt.savefig('eda_results/6_top_scatter_plots.png', dpi=300)
    plt.show()
    
    print("EDA分析完成，图表已保存至 eda_results 文件夹。")

# =============================================================================
# 第二部分：高级特征工程 (Advanced Feature Engineering)
# =============================================================================
def advanced_feature_engineering(data, target_col='Tm of MD (K)'):
    """
    执行特征交互、多项式变换和统计聚合
    :return: X_extended (numpy array), y (numpy array), extended_names (list)
    """
    print("\n=== 执行高级特征工程 ===")

    # 1. 数据准备
    # 假设第3列开始是特征 (索引从0开始，所以是2:)
    # 请根据实际情况修改 data.iloc[:, 2:]
    X_raw = data.drop(columns=[target_col]).iloc[:, 2:].values
    y = data[target_col].values
    feature_names = data.drop(columns=[target_col]).iloc[:, 2:].columns.tolist()
    
    # 处理NaN: 简单的均值填充，防止后续计算报错
    if np.isnan(X_raw).any():
        print("检测到缺失值，正在进行简单均值填充...")
        col_mean = np.nanmean(X_raw, axis=0)
        inds = np.where(np.isnan(X_raw))
        X_raw[inds] = np.take(col_mean, inds[1])

    # 2. 筛选用于交互的Top特征 (避免特征爆炸)
    # 计算每个特征与y的相关性
    correlations = []
    for i in range(X_raw.shape[1]):
        if np.std(X_raw[:, i]) == 0: # 常数列
            correlations.append(0)
        else:
            corr, _ = pearsonr(X_raw[:, i], y)
            correlations.append(abs(corr))
            
    # 选择相关性最高的Top 5特征进行交互
    n_top = min(5, len(feature_names))
    top_indices = np.argsort(correlations)[-n_top:]
    print(f"选择用于交互生成的Top {n_top}特征: {[feature_names[i] for i in top_indices]}")

    # 3. 生成新特征列表
    new_features = []
    new_feature_names = []

    # A. 交互特征 (乘法 & 除法)
    for i in range(len(top_indices)):
        for j in range(i + 1, len(top_indices)):
            idx1, idx2 = top_indices[i], top_indices[j]
            
            # 乘法
            inter_mult = X_raw[:, idx1] * X_raw[:, idx2]
            new_features.append(inter_mult)
            new_feature_names.append(f"{feature_names[idx1]}*{feature_names[idx2]}")
            
            # 除法 (稳健处理除零)
            with np.errstate(divide='ignore', invalid='ignore'):
                # 如果分母绝对值太小，结果置0
                inter_div = np.where(np.abs(X_raw[:, idx2]) > 1e-6, X_raw[:, idx1] / X_raw[:, idx2], 0)
            new_features.append(inter_div)
            new_feature_names.append(f"{feature_names[idx1]}/{feature_names[idx2]}")

    # B. 多项式特征 (平方)
    for idx in top_indices:
        poly = X_raw[:, idx] ** 2
        new_features.append(poly)
        new_feature_names.append(f"{feature_names[idx]}^2")

    # C. 统计特征 (行统计)
    # 仅当特征物理意义相近时才有意义(例如都是原子属性)。如果差异大，可注释掉此段。
    if X_raw.shape[1] > 1:
        new_features.append(np.mean(X_raw, axis=1))
        new_feature_names.append("mean_features")
        
        new_features.append(np.std(X_raw, axis=1))
        new_feature_names.append("std_features")
        
        new_features.append(np.max(X_raw, axis=1))
        new_feature_names.append("max_features")
        
        new_features.append(np.min(X_raw, axis=1))
        new_feature_names.append("min_features")

    # 4. 合并所有特征
    if new_features:
        X_new = np.column_stack(new_features)
        X_extended = np.column_stack([X_raw, X_new])
        extended_names = feature_names + new_feature_names
    else:
        X_extended = X_raw
        extended_names = feature_names

    print(f"原始特征数: {X_raw.shape[1]}")
    print(f"新增特征数: {len(new_feature_names)}")
    print(f"总特征数: {X_extended.shape[1]}")
    
    # 简单检查是否有无穷大值 (除法导致)
    if not np.isfinite(X_extended).all():
        print("警告: 新特征中包含无穷大值，正在替换为0...")
        X_extended[~np.isfinite(X_extended)] = 0

    return X_extended, y, extended_names

# =============================================================================
# 主程序执行入口
# =============================================================================
if __name__ == "__main__":
    # 1. 加载数据
    # ---------------------------------------------------------
    # 场景A: 使用模拟数据 (为了演示代码可以直接运行)
    print("正在生成模拟数据以演示代码...")
    np.random.seed(42)
    N = 200
    df = pd.DataFrame({
        'Compound': [f'Mat_{i}' for i in range(N)], # 第1列：ID
        'Formula': [f'AB_{i}' for i in range(N)],   # 第2列：公式
        'Density': np.random.normal(5, 1, N),       # 特征1
        'Porosity': np.random.uniform(0, 0.5, N),   # 特征2
        'Atomic_Mass': np.random.normal(50, 10, N), # 特征3
        'Hardness': np.random.normal(20, 5, N),     # 特征4
        'Elastic_Mod': np.random.normal(200, 30, N),# 特征5
        'Tm of MD (K)': np.random.normal(3000, 500, N) # 目标变量
    })
    # 添加一些非线性关系
    df['Tm of MD (K)'] += 50 * df['Density']**2 + 100 * df['Hardness'] * df['Porosity']
    
    # 场景B: 如果您有真实文件，请取消下面注释并修改文件名
    # df = pd.read_excel('your_data.xlsx') 
    # ---------------------------------------------------------

    # 2. 运行 EDA
    enhanced_eda(df, target_col='Tm of MD (K)')

    # 3. 运行特征工程
    X_final, y_final, feat_names_final = advanced_feature_engineering(df, target_col='Tm of MD (K)')

    # 4. 结果展示
    print("\n特征工程结果预览 (前5个样本的最后5个特征):")
    print(pd.DataFrame(X_final, columns=feat_names_final).iloc[:5, -5:])
    
    print("\n代码执行完毕。")
```
